 # Student Marks Analyzer

Create a student mark analyzer that can store the marks of students (according to the user's wish) and find the total marks, average marks, highest marks, lowest marks and assign grade on the basis of average.

# Logic

We first create a user defined function which sets the boundaries for the grade every average marks should get. Then we initialize a list, marks_list to enable us to store values inside it. Before we start the while loop, we declare the default value of continue_entry, which is yes.

We will then start the while loop and set it so that as long as the answer to continue_entry is "yes" or "YES" (.lower()) is hence used, the loop will continue as long as the user's wish. The variable score takes the input from the user and appends it into the marks_list which we initialized earlier.

Once the user has completed all entries, the interpreter gets out of the loop and then we convert the list into an array marks_array with np.array(marks_list). We do it so that our operations (finding highest, lowest, average, etc.) can be done quite simply without many lines of code.

Then, as said above, the total, average, highest and lowest marks of the students are displayed in the form of "Summary" and the grade is displayed according the student's average mark according the boundaries set in the user defined function we created at the beginning.



```python
import numpy as np

# User defined function to set the grading of the marks
def grade(average):
    if average > 85:
        return "A"
    elif average >= 70:
        return "B"
    elif average >= 50:
        return "C"
    else:
        return "Fail"

# Initializing variables
marks_list = []
continue_entry = "yes"
# While loop to continue entries till the answer is yes
while continue_entry == "yes":
    score = float(input("Enter student marks: "))
    # Adding the score to the marks list using append()
    marks_list.append(score)

    # Ask the user if they want to keep going
    # If they type anything other than 'yes', the loop will stop
    continue_entry = input("Do you want to enter more marks of students? (yes/no): ").lower()
    # .lower() in case they type "Yes" or "YES"

# Converting the list into a Numpy array
marks_array = np.array(marks_list)

# Using the functions to find out the total, average, highest and lowest of the marks entered
total = np.sum(marks_array)
average = np.mean(marks_array)
highest = np.max(marks_array)
lowest = np.min(marks_array)

# Get the final grade of all the students from the function grade(average)

final_grade = grade(average)

# Displaying the results
print("Project Analysis Summary: ")
print("Total Marks:", total)
print("Average Marks:", average)
print("Highest Marks:", highest)
print("Lowest Marks:", lowest)
print("Grade Based on Average:", final_grade)
```
